```python
import numpy as np

x=[1,2,9]
y=[3,5,7]
x1=[3,6,8]
y1=[4,8,12]

plt.title("simple")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.bar(x,y,label="water",color="g")
plt.bar(x1,y1,label="soil",color="r")
plt.legend()
plt.grid(color='y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x2=np.arange(0,100,0.5)
y2=[i for i in range(len(x2))
plt.title("simple")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.bar(x2,y2,label="water",color="g")
plt.legend()
plt.grid(True,color='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.random.normal(size = 500) #use of this function 
plt.hist(y, bins=30)
plt.ylabel('Probability')
plt.show()
```
